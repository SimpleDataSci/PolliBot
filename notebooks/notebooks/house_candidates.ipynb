```python
from urllib import request, parse
from bs4 import BeautifulSoup
import pandas as pd
import wikipedia
import os
import re
```

```python
zipcodes = pd.read_csv('us-zipcode.csv', low_memory=False)
```

```python
zipcodes.head()
```

```python
states = zipcodes.State.unique()
```

```python
for state in states:
    url = 'http://www.ontheissues.org/{}_{}.htm'.format('House', state)
    raw_html = open('{}/{}_{}_html.txt'.format('house', 'House', state),
                            'w')
    try:
        headers = {}
        headers['User-Agent'] = """
        Mozilla/5.0 (X11; Linux i686) AppleWebKit/537.17 (KHTML, like Gecko)
        Chrome/24.0.1312.27 Safari/537.17"""

        req = request.Request(url, headers=headers)
        resp = request.urlopen(req)

        print(url)
        respData = resp.read()
        raw_html.write(str(respData))
        raw_html.close()
        raw_html = open('{}/{}_{}_html.txt'.format('house', 'House', state),
                            'r').read()
        
        html = BeautifulSoup(raw_html, 'html.parser')
        for script in html(["script", "style"]):
            script.decompose()  # rip it out
        
        with open('{}/{}_{}.txt'.format('house', 'House', state), 'w') as save_file:
            for p in html.select('li'):
                name = re.sub(r'\\[a-z]', '', p.text)

                # 'name', 'topic', 'comment', 'state', 'election', 'bio', 'link'
                save_file.write(str(name))
                print(name + '\n')
    except Exception as e:
        raw_html.close()
        with open('{}/{}_{}.txt'.format('house', 'House', state), 'w') as save_file:
            print(str(e))
```

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
for state in states:
    try:
        with open('{}/{}_{}.txt'.format('house', 'House', state), 'r') as save_file:
            temp = ''
            cont = False
            for f in word_tokenize(save_file.read()):
                if cont or f == '(' or f == ')':
                    temp += f
                    cont = True
                print(f)
                print(temp)
                temp = ''
                cont = False
    except Exception as e:
        print(str(e))
```
